# Limpieza de datos del DataFrame de Chartmasters

Extraido de: https://chartmasters.org/best-selling-artists-of-all-time/

```python
import pandas as pd

bsat = pd.read_csv('..\data\Best_Sellings_All_Time.csv', sep='|', index_col=0)
bsat
```

```python
# Comenzaremos eliminando algunas columnas que no nos aportan información útil, como 'Sales Update' y 'Streams EAS (Update)'"RIAA Jupyter.ipynb"

bsat.drop('Sales Update', axis=1, inplace=True)
bsat.drop('Streams EAS (Update)', axis=1, inplace=True)
bsat
```

```python
# Vemos que los numeros del df son enteros pero estan separados por comas, lo que nos indica que son strings. Los cambiaremos a enteros.
for col in bsat.columns[1:]:
    if bsat[col].dtype == 'object':
        bsat[col] = bsat[col].replace(',', '', regex=True).astype(int)
```

```python
'''
Hemos sacado esta tabla haciendo web scraping y la tenemos bastante 'limpia'. De aquí podemos sacar distinta información útil:
- Esta ordenada de menor a mayor según la escala CSPC (Commensurate Sales to Popularity Concept) que se basa en diferentes valores de ventas.
- Esto nos indica que en esta tabla todos los artistas son distintos y no se repite ninguno, por lo que el parametro top de la columna 'Artist' coincidirá con el primer elemento de ésta, The Beatles.
- Mismo caso con la columna 'Total CSPC', puesto que no hay valores repetidos, el parametro top coincide con el primer elemento.
- Mismo caso para la columna 'Studio Album Sales'.
- En la columna de 'LPs Sales', el valor que más se repite es el 0, un total de 8 veces. Este valor será interesante de analizar su por qué.
''';
```

```python
# Visualización de los datos

import seaborn as sns
import matplotlib.pyplot as plt

bsat_25 = bsat.head(25)

# Configurar la visualización
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Crear la gráfica de barras apiladas
sns.barplot(y="Artist", x="Total CSPC", data=bsat_25, color="deepskyblue")
sns.barplot(y="Artist", x="Studio Albums Sales", data=bsat_25, color="pink") 
sns.barplot(y="Artist", x="Other LPs Sales", data=bsat_25, color="orange")
sns.barplot(y="Artist", x="Physical Singles Sales", data=bsat_25, color="purple")
sns.barplot(y="Artist", x="Digital Singles Sales", data=bsat_25, color="darkblue")

# Añadir título y etiquetas de los ejes
plt.title("Ventas de artistas según tipo de medio")
plt.xlabel("Artistas")
plt.ylabel("Ventas totales (CSPC)")

# Ajustar la leyenda
legend = plt.legend(['Total CSPC', 'Studio Albums Sales', 'Other LPs Sales', 
                     'Physical Singles Sales', 'Digital Singles Sales'], loc='lower right')


# Mostrar la visualización
plt.show()
```
